```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy import stats as sc_st
import pandas as pd
import math
```

Question 1

```python
j=1
for N in (1,5,10):
  f = plt.figure()
  f.set_figwidth(5)
  f.set_figheight(7.5)
  plt.subplot(3,1,j)
  x=[]
  for i in range(1000):
    sample=np.random.chisquare(3,N)
    x.append(np.mean(sample))
  sig=(6/N)**0.5
  dist=sc_st.norm(3,sig)
  x1=np.linspace(0,10)
  y=dist.pdf(x1)
  plt.plot(x1,y)
  plt.hist(x,density=True)
  i+=1
  plt.show()
```

Question 2

```python
ls=pd.read_table("http://www.iith.ac.in/~shantanud/test.dat",sep="\s+")
l=ls.to_numpy()
x=list()
y=list()
for i in range(len(l)):
  x.append(l[i][0])
  y.append(l[i][1])
plt.figure()
plt.scatter(np.log10(y),np.log10(x))
plt.xlabel("Redshift(log)")
plt.ylabel("Luminosity(log)")
plt.show()
print(sc_st.kendalltau(y,x))
print(sc_st.spearmanr(y,x))
print("Pearson Result(corelation=",sc_st.pearsonr(y,x)[0],", pvalue=",sc_st.pearsonr(y,x)[1],")")
print("The is a moderate co-relation as luminosity increases with redshift")
```

```python
sample=list()
n=1000
sample.append(np.random.normal(size=n,loc=0, scale=1))
sample.append(np.random.normal(size=n,loc=0, scale=1))
pc=sc_st.pearsonr(sample[0],sample[1])[0]
pv=sc_st.pearsonr(sample[0],sample[1])[1]
print("Pearson Result(corelation=",pc,", pvalue=",pv,")")
t=pc*math.sqrt((n-2)/(1-pc**2))
if t>=0:
  p=2*(1-sc_st.t.cdf(t,n))
else:
  p=2*(sc_st.t.cdf(t,n))
print("P-value of Student-t Distribution=",p)
if (np.allclose(pv,p)):
  print("Both p-values are similar")
else:
  print("Both p-values are not similar")
```

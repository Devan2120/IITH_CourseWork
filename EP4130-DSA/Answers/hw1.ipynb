```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
from scipy import stats as sc_st
from astroML import stats as astro_st
```

QUESTION 1

```python
sample=np.random.normal(size=1000,loc=1.5, scale=0.5)
x=np.linspace(0,3,num=32)
prob=np.zeros(len(x))
for i in sample:
  for j in range(len(x)-1):
    if x[j]<=i and i<=x[j+1]:
      prob[j]+=1
prob/=1000
plt.bar(x,prob)
plt.show()
MAD=sc_st.median_abs_deviation(sample)
sig=astro_st.sigmaG(sample)

print("Mean:",np.mean(sample))
print("Variance:",np.var(sample))
print("Skewness:",sc_st.skew(sample))
print("Kurtosis:",sc_st.kurtosis(sample))
print("MAD:",MAD)
print("Standard deviation:",1.482*MAD)
print("Sigma_g:",sig)
```

QUESTION 2

```python
mean=0
x=np.linspace(-5,5)
gam=1.5
dist1=sc_st.cauchy(mean,gam)
dist2=sc_st.norm(mean,gam)
y1=dist1.pdf(x)
y2=dist2.pdf(x)
plt.plot(x,y1,linestyle="dotted")
plt.plot(x,y2,linestyle="dashed")
plt.legend(["cauchy","Guassion"])
plt.show()
```

QUESTION 3

```python
x=np.arange(-100,100)
dist3=sc_st.poisson(5)
dist2=sc_st.norm(5,np.power(5,0.5))
y3=dist3.pmf(x)
y2=dist2.pdf(x)
plt.plot(x,y3)
plt.scatter(x,y2,color="green")
plt.legend(["poisson","normal"])
plt.show()
```

```python
life=[0.8920,0.881,0.8913,0.9837,0.895]
err=[0.00044,0.009,0.00032,0.00048,0.00045]
sig=np.power(err,-2)
var=np.sum(sig)
mean=np.sum(life*sig)
mean/=var
print("Weighted Mean:",mean)
print("Uncertainity:",np.power(1/var,0.5))
```
